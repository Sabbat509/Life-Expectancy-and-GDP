```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("all_data.csv")
```

```python
data
```

```python
sns.set(style="whitegrid")
```

# Life expectancy at birth (years)

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country', marker='o')
plt.title('Life Expectancy Over Time in Six Nations')
plt.ylabel('Life Expectancy at Birth (years)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.show()
```

# GDP over time for each country

```python
# Create a plot for GDP over time for each country
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Over Time in Six Nations')
plt.ylabel('GDP (in trillions)')
plt.xlabel('Year')
plt.legend(title='Country')
plt.show()
```

# Correlation between GDP and Life Expectancy

```python
# Create a scatter plot to visualize the correlation between GDP and life expectancy
plt.figure(figsize=(12, 6))
sns.scatterplot(data=data, x='GDP', y='Life expectancy at birth (years)', hue='Country')
plt.title('Correlation between GDP and Life Expectancy')
plt.ylabel('Life Expectancy at Birth (years)')
plt.xlabel('GDP (in trillions)')
plt.legend(title='Country')
plt.show()

# Calculate the correlation coefficient
correlation = data[['GDP', 'Life expectancy at birth (years)']].corr().iloc[0, 1]
print('Correlation coefficient:', correlation)
```

# Average Life Expectancy

```python
average_life_expectancy = data['Life expectancy at birth (years)'].mean()
print('Average Life Expectancy:', average_life_expectancy)
```

# Distribution of Life Expectancy

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Life expectancy at birth (years)'], kde=True)
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.show()
```

# Life Expectancy and GDP Over Time for Each Country

```python
countries = data['Country'].unique()

for country in countries:
    country_data = data[data['Country'] == country]

    # Life Expectancy Over Time
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='Year', y='Life expectancy at birth (years)', marker='o')
    plt.title(f'Life Expectancy Over Time in {country}')
    plt.ylabel('Life Expectancy at Birth (years)')
    plt.xlabel('Year')
    plt.show()

    # GDP Over Time
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=country_data, x='Year', y='GDP', marker='o')
    plt.title(f'GDP Over Time in {country}')
    plt.ylabel('GDP (in trillions)')
    plt.xlabel('Year')
    plt.show()
    
     # Scatter plot of GDP vs Life Expectancy
    plt.figure(figsize=(12, 6))
    sns.scatterplot(data=country_data, x='GDP', y='Life expectancy at birth (years)')
    plt.title(f'Correlation between GDP and Life Expectancy in {country}')
    plt.ylabel('Life Expectancy at Birth (years)')
    plt.xlabel('GDP (in trillions)')
    plt.show()

    # Calculate the correlation coefficient
    correlation = country_data[['GDP', 'Life expectancy at birth (years)']].corr().iloc[0, 1]
    print(f'Correlation coefficient in {country}:', correlation)
    
    average_life_expectancy = country_data['Life expectancy at birth (years)'].mean()
    print(f'Average Life Expectancy in {country}:', average_life_expectancy)
    
     # Distribution of Life Expectancy
    plt.figure(figsize=(10, 6))
    sns.histplot(country_data['Life expectancy at birth (years)'], kde=True)
    plt.title(f'Distribution of Life Expectancy in {country}')
    plt.xlabel('Life Expectancy at Birth (years)')
    plt.ylabel('Frequency')
    plt.show()
```
